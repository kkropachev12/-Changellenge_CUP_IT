```python
# Импортируем библиотеки
import os
import pandas as pd
```

```python
clusters = pd.read_csv('clusters.csv', sep=',')
clusters = clusters.set_index('qw')
clusters = clusters.drop(columns=['Unnamed: 0'])

for pay_type in ['payin', 'payout']:
    for file in os.listdir(f'categories/{pay_type}'):
        print(file)
        df = pd.read_csv(f'categories/{pay_type}/{file}', sep=',')
        df = df.set_index('qw')
        df = df.drop(columns=['Unnamed: 0'])
        df = pd.concat([df, clusters], axis=1)
        df = df.dropna()
        # Recency
        day = '2022-01-01' # Начальная дата - самая ранняя
        day = pd.to_datetime(day)
        df['date_max'] =  pd.to_datetime(df['date_max']).dt.normalize()
        df['date_min'] =  pd.to_datetime(df['date_min']).dt.normalize()
        dates = df['date_max'].to_list()
        df = df.assign(date = -1 * (day - df['date_max'])) # Берем начальную и вычитаем из нее максимальную дату 
        df['date'] = [-1*(day - date).days for date in dates]
        # LT
        # df['LT'] = df.assign(LT=(df['date_max'] - df['date_min']))
        df.to_csv(f'categories_clusters/{pay_type}/{file}')
        group = df.groupby(['Clusters']).agg({
                                'qw_count': 'sum',
                                'amount': 'sum',
                                'revenue': 'sum',
                                'costs': 'sum',
                                'date': 'mean'
                            })
        group = group.assign(average_check=group['amount']/group['qw_count'])
        group = group.assign(margin=group['revenue']/group['amount'])

        group.to_csv(f'categories_clusters_mean/{pay_type}/{file}')
```
