```python
import os
import pandas as pd

def get_data(df):
    '''Выбирает нужные столбцы с данными'''
    df.dropna()
    qw_count = df.groupby(['qw'])['qw'].count()

    result = pd.DataFrame()
    result['qw_count'] = qw_count
    return result

def join_dfs(df1, df2):
    df = df1.append(df2)
    df.reset_index(inplace=True)
    df = df.groupby(['qw']).agg({
        'qw_count': 'sum'
    })
    return df
```

```python
def group_cats(df):
    '''Убирает в столбце категорий цифры'''
    df['cat'] = df['cat'].astype('str')
    cats = df['cat'].to_list()
    cats_new = []
    for cat in cats:
        if cat[-1].isdigit():
            cat = cat[:-2].strip()
        cats_new.append(cat)
    df['cat'] = cats_new
    return df
```

```python
def spliter_in_out(df):
    '''Делит датафрейм на два payin и payout'''
    in_out_dict = {}
    df = group_cats(df)
    for pay_type in ['payin', 'payout']:
        in_out_dict[pay_type] = df.loc[df['txn_type'] == pay_type]
    return in_out_dict
```

```python
def spliter_cats(in_out_dict):
    '''Распределяет в payin и payout данные по категориям'''
    in_out_category_dict = {'payin': {}, 'payout': {}}
    for pay_type, df in in_out_dict.items():
        cats = list(df['cat'].unique())
        for cat in cats:
            in_out_category_dict[pay_type][cat] = df.loc[df['cat'] == cat]
    return in_out_category_dict
```

```python
def main(df, in_out_categories):
    df = group_cats(df)
    in_out_dict = spliter_in_out(df)
    in_out_category_dict = spliter_cats(in_out_dict)
    for pay_type, cats in in_out_category_dict.items():
        for cat, df in cats.items():
            result = get_data(df)
            try:
                df1 = in_out_categories[pay_type][cat]
                in_out_categories[pay_type][cat] = join_dfs(df1, result)
            except:
                in_out_categories[pay_type][cat] = result
    return in_out_categories
```

```python
in_out_categories = {'payin': {}, 'payout': {}}
for file in os.listdir('dataset/txns_data_raw'):
    print(file)
    print(len(in_out_categories['payin']))
    df = pd.read_csv(f"dataset/txns_data_raw/{file}", sep=';')
    main(df, in_out_categories)

for pay_type, cats in in_out_categories.items():
    for cat, df in cats.items():
        cat = cat.replace('/', '')
        df.to_csv(f'categories_correct/{pay_type}/{cat}.csv')
```
